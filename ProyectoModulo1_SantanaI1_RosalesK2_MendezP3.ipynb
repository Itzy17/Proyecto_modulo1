## 1. Accidentes de tránsito.

### 1.1 Accidentes de tránsito terrestre en zonas urbanas y suburbanas.
> En este trabajo investigaremos los accidentes de tránsito que se han presentado en las zonas urbanas en Jalisco desde 2008 hasta 2018, creando simulaciones para observar el número de eventos y así, intentar de disminuir dichos accidentes y concientizar a las personas sobre la gravedad que conlleva el conducir sin estar en nuestros 5 sentidos y prestando atención a todo nuestro alrededor.



### 1.2 Objetivos.

> #### 1.1 Objetivo general.
>  - Analizar la cantidad de accidentes automovilísticos que ocurren en Jalisco y estimar una reducción de un 3% en los sucesos.

> #### 1.2 Objetivos específicos
>>  - Evaluar el porcentaje de choques entre hombres y mujeres.
>>  - Comparar la cantidad de accidentes en mujeres menores de edad contra mujeres mayores de edad.
>>  - Comparar la cantidad de accidentes en hombres menores de edad contra hombres mayores de edad.
>>  - Valorar la cantidad de decesos y personas con vida entre hombres y mujeres. 



### 1.3 Definición del problema.

> "Jalisco es el primer lugar en accidentes de tránsito"

> Actualmente hay una gran gama de problemas que existen a nivel mundial, continental, nacional, estatal, local, familiar, y personal. Sin embargo la razón por la que decidimos centrarnos en el tema: ACCIDENTES AUTOMOVILÍSTICOS, es debido a que las estadísticas de accidentes automovilísticos son alarmantes. Los Accidentes automovilísticos se deben a la combinación de diversos factores. Y, si se pretende solucionar el problema, es básico que se conozca las causas y consecuencias del problema, y una vez conociendo las características de la problemática podemos proponer más y mejores soluciones.

> Es muy triste que a pesar de los avances de la medicina, y tecnología en los autos, las muertes por accidentes automovilísticos se hayan convertido en uno de los más graves problemas de salud pública.

> Este problema, es un asunto que a todos nos concierne, pues todos hemos experimentado este problema ya sea en nuestra persona, o algún conocido.

> Se han llevado acabo diversas investigaciones sobre accidentes automovilísticos, sin embargo, no hemos logrado solucionar este problema. No podemos permitir que las cifras de accidentes automovilísticos sigan creciendo, es por ello que decidimos hablar sobre este tema, para poder brindar información y por supuesto poder plantear posibles soluciones.

> Como hemos visto en las noticias de los últimos años cada vez se vuelven más recurrentes los accidentes automovilísticos, con el paso del tiempo la gravedad y la incurrencia de este suceso ha ido en aumento. Hemos visto que los motivos por los cuales se presentan los accidentes pueden ser simples descuidos en los cuales muchas veces termina en solo un impacto o por otra parte, acabar en una desgracia.

> En este proyecto se harán simulaciones enfocándonos en las principales preguntas que se vienen a tu mente cuando ocurre un accidente automovilístico, entonces, nuestros 4 nodos serán:

>> a) **Las víctimas que están involucradas**, es decir, número de personas fallecidas y/o heridas que iban en los respectivos autos.

>> b) **Los autos que están involucrados**, será el número de autos que estuvieron en el percance, especificando que tipo de vehículo estuvo involucrado, por ejemplo: microbús, bicicleta, camión de carga, entre otros.

>> c) **Sexo**, en este apartado especificaremos la cantidad de hombres y/o de mujeres que se presentaron en el desastre. En caso de que alguna persona se haya fugado existe un apartado en el cual se mostrará la cantidad de vehículos que se dieron a la fuga.

>> d) **Edad**, en este último nodo se darán a conocer si los participantes eran menores o mayores de edad.

> El objetivo principal de este proyecto es ofrecer al lector información sobre los accidentes automovilísticos con altos índices de muertes y lo que es más alarmante, que los adolescentes y adultos jovenes son los más involucrados en ellos.

> Así como también por medio de este trabajo se planea conocer todas las causas de los accidentes y de alguna forma tratar de evitarlos. A través de esta investigación se pretende evitar a toda costa que estos terribles accidentes sigan sucediendo, realizando campañas que estén dirigidas a los jóvenes de preparatoria y a los jóvenes universitarios, para sí poder reducir el índice de los accidentes producidos por éstos.

> En conclusión, muchos de los accidentes que ocurren los podríamos evitar sencillamente prestando un poco más de atención a nuestro alrededor, porque muchas veces no es que tu conduzcas mal o no pongas suficiente atención, más bien es que la otra persona no estaba 100% concentrada sobre los movimientos que se hacen al conducir.



### 1.4 Nodos y variables que se decidieron simular y porqué
> Las variables que vamos a abordar en este proyecto son:

- Las víctimas que están involucradas.
- Los autos que están involucrados.
- Sexo 
- Edad

Decidimos simular estos nodos porque cuando ocurre algún accidente automovilístico, lo primero que se preguntan las personas tanto civiles como autoridades, es el número de personas, sexo y edad.
